# Q13. Concatenate two DataFrames vertically.

```python
# YOUR CODE HERE

Table1 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bangalore']}
df1 = pd.DataFrame(Table1)
print(df1)
print("This is Table1 +++++++++++++++++++++++/" )
Table2 = {
    'Name': ['NKS', 'Bobu', 'Charly' ],
    'Age': [26, 25, 45],
    'City': ['Lucknow', 'MP', 'Andhra']}
df2 = pd.DataFrame(Table2)
print(df2)
print("This is Table2 +++++++++++++++++++++++/" )
concatenated_df = pd.concat([df1, df2], ignore_index=True)
print("This is Merged data of the Table1 & Table2 ++++++++++++++++++++/" )
print(concatenated_df)
#raise NotImplementedError()
```
